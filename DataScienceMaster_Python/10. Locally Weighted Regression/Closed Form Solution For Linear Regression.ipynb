```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
# GENERATE DATASET
X,Y=make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)
Y=Y.reshape((-1,1))
X.shape,Y.shape
```

```python
# Normalise
X=(X-X.mean())/X.std()

# Visualise
plt.figure()
#plt.style.use('seaborn')
plt.scatter(X,Y)
plt.title("Normalised Data")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
ones=np.ones((X.shape[0],1))
X_=np.hstack((X,ones))
print(X_.shape)
print(X_[:4,])
```

```python
def predictions(X,theta):
    return np.dot(X,theta)

def getTheta_closedForm(X,Y):
    Y=np.mat(Y)
    firstPart=np.dot(X.T,X)
    secondPart=np.dot(X.T,Y)
    # using the linear algebra psudo inverse method
    theta=np.linalg.pinv(firstPart)*secondPart
    return theta
```

```python
theta=getTheta_closedForm(X_,Y)
print(theta)
```

```python
# Visualise
plt.figure()
#plt.style.use('seaborn')
plt.scatter(X,Y)
plt.plot(X,predictions(X_,theta),c='Black',label='Prediction')
plt.legend()
plt.title("Normalised Data")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
