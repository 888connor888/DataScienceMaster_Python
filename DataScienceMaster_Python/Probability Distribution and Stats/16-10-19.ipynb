# Multivariate Normal Distribution


```python
# Here we are talking about a bi-variate case
mean_vector=np.array([0.0,0.0])#centered around the origin
coverient_matrix=np.array([[1,0.8],[0.8,1]])
mean_vector2=np.array([5.0,5.0])
coverient_matrix2=np.array([[5,0.8],[0.8,2]])
dist=np.random.multivariate_normal(mean_vector,coverient_matrix,500)# 500 is the number of points
dist2=np.random.multivariate_normal(mean_vector2,coverient_matrix2,500)
```

```python
print(dist)
```

```python
print(dist.shape)
```

```python
plt.scatter(dist[:,0],dist[:,1])# the x cood is zeroth coumn of the matrix and y cood as first
plt.scatter(dist2[:,0],dist2[:,1])
plt.style.use('seaborn')
plt.show()
```
