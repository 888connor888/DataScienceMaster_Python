# Bar Graphs

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(10)
y=np.arange(10)
plt.bar(x,y)
plt.show()
```

```python
plt.style.use('seaborn')
x_coordinates=np.array([0,1,2])*2
plt.bar(x_coordinates,[10,20,30],width=.5,label='current',tick_label=['Gold','Platinum','silver'])
plt.bar(x_coordinates+.5,[20,10,12],width=.5,label='next')
plt.title('plot of metal comparision')
plt.ylim(0,40)
plt.legend()
plt.show()
```

### Standard normal distribution(mean=0,standard deviation is =1)

```python
xsn=np.random.randn(100)
sigma=5
u=60
x=xsn*sigma+u
x2=xsn*6+40
print(xsn)
```

```python
print(x)
```

```python
x=np.round(x)
print(x)
```

```python
plt.hist(x)
plt.hist(x2,alpha=.8)
plt.xlabel('Marks Range')
plt.ylabel('Prob/Freqcount of Students')
plt.show()
```
