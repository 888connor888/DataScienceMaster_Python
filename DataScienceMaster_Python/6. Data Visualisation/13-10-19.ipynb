```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line Plot

```python
plt.style.use('seaborn')
x=np.arange(10)
print(x)
y1=x**2
y2=2*x+3
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='red',label='apple')
plt.xlabel('Time')
plt.ylabel('price')
plt.title('prices of fruits')
plt.legend()
plt.show()# the plot is no longer an object it is actually a graph now
```
