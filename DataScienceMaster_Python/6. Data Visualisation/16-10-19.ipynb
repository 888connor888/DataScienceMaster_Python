```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("movie_metadata.csv")
```

```python
df.head(n=10)
```

```python
df.columns
```

```python
titles=list(df.get('movie_title'))
```

```python
titles[:5]
```

```python
freq_titles={}
for title in titles:
    length=len(title)
    if freq_titles.get(length) is None:
        freq_titles[length]=1
    else:
        freq_titles[length]+=1
```

```python
print(freq_titles)
```

```python
X=np.array(list(freq_titles.keys()))
Y=np.array(list(freq_titles.values()))
```

```python
plt.scatter(X,Y)
plt.xlabel('Length of Characters')
plt.ylabel('Freq Count')
plt.title('Movie Data Visualisation')
plt.style.use('seaborn')
plt.show()
```
