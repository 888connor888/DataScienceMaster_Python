```python
from urllib.request import urlopen
```

```python
X="https://github.com/coding-blocks-archives/machine-learning-online-2018/blob/master/Datasets/Linear%20Regression/linearX.csv"
```

```python
X_data=urlopen(X)
print(type(X_data))
```

```python
X_data_html=X_data.read()
print(X_data_html)
X_data.close()
```

```python
!pip3 install bs4
```

```python
from bs4 import BeautifulSoup as soup
```

```python
X_soup=soup(X_data_html,'html.parser')
```

```python
print(X_soup)
```

```python
table_rows=X_soup.findAll('tr')
```

```python
print(table_rows)
```

```python
Table_rows=[]
c=1
for row in table_rows:
    current_row=[]
    row_data=row.findAll('td')[1:]
    
    for idx,data in enumerate(row_data):
        current_row.append(str(c))
        current_row.append(data.text)
    c+=1    
    Table_rows.append(current_row)    
    
```

```python
print(Table_rows)
```

```python
filename="linearX.csv"
```

```python
with open(filename,'w') as f:
    for row in Table_rows:
        row_string=''
        for w in row:
            row_string=w    
        row_string+='\n'
        f.write(row_string)
```
