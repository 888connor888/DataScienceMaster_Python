```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx=pd.read_csv('xdata.csv')
dfy=pd.read_csv('ydata.csv')
```

```python
# converting datasets as numpy array
X= dfx.values
Y= dfy.values
X=X[:,1:]
Y=Y[:,1:].reshape(-1,)
```

```python
# This is the Euclidian Distance Function
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
def knn(X,Y,queryPoint,k=5):
    vals=[]
    m=X.shape[0]
    for i in range(m):
        d=dist(queryPoint,X[i])
        vals.append((d,Y[i]))
    vals=sorted(vals)
    vals=vals[:k]
    vals=np.array(vals)
    new_vals=np.unique(vals[:,1],return_counts=True)
    print(new_vals)
    index=new_vals[1].argmax()
    print(index)
    predictions=new_vals[0][index]
    print(predictions)
    return predictions
```

```python
s=knn(X,Y,[2,3])
```
